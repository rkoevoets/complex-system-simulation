```python
import networkx as nx
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from main_numpy import NeuronModel
import powerlaw 
```

```python
er_network = nx.erdos_renyi_graph(400, 0.1, directed=False)
er_network.add_node(-1)  # Adding the sink node

p = 0.9  # Probability of adding an edge to the sink

# Iterate over each node (not the sink node)
for node in er_network.nodes():
    if node != -1:
        if random.random() < p:
            er_network.add_edge(node, -1)

er_network_model = NeuronModel(er_network)   
er_network_data = np.array(er_network_model.run(100000))

```

```python
er_network_data
```

```python
# Calculate the unique avalanche sizes and their frequencies
avalanche_sizes_er, frequencies_er = np.unique(er_network_data, return_counts=True)

# Plot the data points on a log-log scale
plt.figure()
plt.scatter(avalanche_sizes_er, frequencies_er)

# Set log-log scale
plt.xscale('log')
plt.yscale('log')

# Add labels and title
plt.xlabel('Avalanche Size (log)')
plt.ylabel('Frequency (log)')
plt.title('Frequency Distribution of Avalanche Sizes Random Network')

# Show the plot
plt.show()
```

```python
# plotting scatter plot
plt.hist(er_network_data, density=False, log=False)

plt.title("Avalanche size distribution")
plt.xlabel("s")
plt.ylabel("P(s)")

plt.show()
```
