```python
import networkx as nx
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from main_numpy import NeuronModel
```

```python
# create watts-strogatz network
ws_network_100 = nx.watts_strogatz_graph(100, 10, 0.1) # 40 nodes, connected to 4 neighborhs, and p=0.2
# add sink 
ws_network_100.add_node(-1)
# connect sink
for node in ws_network_100.nodes():
    if node != -1:
        if random.random() < 0.2:
            ws_network_100.add_edge(node, -1)

ws_model_100 = NeuronModel(ws_network_100)   

ws_data_100 = np.array(ws_model_100.run(50000))  # number of steps

# Calculate the unique avalanche sizes and their frequencies
avalanche_sizes_ws_100, frequencies_ws_100 = np.unique(ws_data_100, return_counts=True)
```

```python
# Plot the data points on a log-log scale
plt.figure()
plt.scatter(avalanche_sizes_ws_100, frequencies_ws_100)

# Set log-log scale
plt.xscale('log')
plt.yscale('log')

# Add labels and title
plt.xlabel('Avalanche Size')
plt.ylabel('Frequency')
#plt.title('Watts-Strogatz: Frequency Distribution of Avalanche Sizes')

# Show the plot
plt.show()
```

```python
def plot_scatter(avalanche_sizes, frequencies, label, color=None):
    plt.scatter(avalanche_sizes, frequencies, s=20, color=color, label=label)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('s')
    plt.ylabel('Frequency')
    plt.legend()
    plt.savefig(f'ws_{label}_scatter.png', dpi=300)
    plt.show()

data = [[avalanche_sizes_ws_100, frequencies_ws_100, 'N=100'],
        [avalanche_sizes_ws_200, frequencies_ws_200, 'N=200'], 
        [avalanche_sizes_ws_600, frequencies_ws_600, 'N=600'],
        [avalanche_sizes_ws_400, frequencies_ws_400, 'N=400'],
        [avalanche_sizes_ws_1600, frequencies_ws_1600, 'N=1600'],
        [avalanche_sizes_ws_3600, frequencies_ws_3600, 'N=3600'], 
        [avalanche_sizes_ws_4000, frequencies_ws_4000, 'N=4000']]

for avalanche_sizes, frequencies, label in data:
    plot_scatter(avalanche_sizes, frequencies, label)
```

```python
# to get exponent by fitting data with large node size
def power_law(x, a, b):
    return a* x**-b  

avalanche_sizes_ws_3600, frequencies_ws_3600 = np.unique(ws_data_4000, return_counts=True)
popt_3600, pcov_3600 = curve_fit(power_law, avalanche_sizes_ws_3600, frequencies_ws_3600)

x_fit_3600 = np.logspace(np.log10(min(avalanche_sizes_ws_3600)), np.log10(max(avalanche_sizes_ws_3600)), 10000)
y_fit_3600 = power_law(x_fit_3600, *popt_3600)
norm_constant_3600 = np.trapz(y_fit_3600, x=x_fit_3600)
pdf_fit_3600 = y_fit_3600 / norm_constant_3600

a_3600, b_3600 = popt_3600
print(b_3600)  ## estimated critical exponent
```

```python
# Define the model function (Power law with exponential cutoff)
def power_law_with_exponential_cutoff(x, a, b, c):
    return a* x**-b * np.exp(-c * x) 

# Perform curve fitting
popt_200, pcov_200 = curve_fit(power_law_with_exponential_cutoff, avalanche_sizes_ws_200, frequencies_ws_200)
popt_400, pcov_400 = curve_fit(power_law_with_exponential_cutoff, avalanche_sizes_ws_400, frequencies_ws_400)
popt_600, pcov_600 = curve_fit(power_law_with_exponential_cutoff, avalanche_sizes_ws_600, frequencies_ws_600)
popt_100, pcov_100 = curve_fit(power_law_with_exponential_cutoff, avalanche_sizes_ws_100, frequencies_ws_100)


# Generate fitted curve values
x_fit_200 = np.logspace(np.log10(min(avalanche_sizes_ws_200)), np.log10(max(avalanche_sizes_ws_200)), 1000)
y_fit_200 = power_law_with_exponential_cutoff(x_fit_200, *popt_200)
x_fit_400 = np.logspace(np.log10(min(avalanche_sizes_ws_400)), np.log10(max(avalanche_sizes_ws_400)), 1000)
y_fit_400 = power_law_with_exponential_cutoff(x_fit_400, *popt_400)
x_fit_600 = np.logspace(np.log10(min(avalanche_sizes_ws_600)), np.log10(max(avalanche_sizes_ws_600)), 1000)
y_fit_600 = power_law_with_exponential_cutoff(x_fit_600, *popt_600)
x_fit_100 = np.logspace(np.log10(min(avalanche_sizes_ws_100)), np.log10(max(avalanche_sizes_ws_100)), 1000)
y_fit_100 = power_law_with_exponential_cutoff(x_fit_100, *popt_100)


# Calculate normalization constant (integral over the defined range)
norm_constant_200 = np.trapz(y_fit_200, x=x_fit_200)
norm_constant_400 = np.trapz(y_fit_400, x=x_fit_400)
norm_constant_600 = np.trapz(y_fit_600, x=x_fit_600)
norm_constant_100 = np.trapz(y_fit_100, x=x_fit_100)

# Normalize the fitted curve to obtain the PDF
pdf_fit_200 = y_fit_200 / norm_constant_200
pdf_fit_400 = y_fit_400 / norm_constant_400
pdf_fit_600 = y_fit_600 / norm_constant_600
pdf_fit_100 = y_fit_100 / norm_constant_100

# Plot the original data and fitted PDF on a log-log scale
plt.loglog(x_fit_100, pdf_fit_100, label='N=100')
plt.loglog(x_fit_200, pdf_fit_200, label='N=200')
plt.loglog(x_fit_400, pdf_fit_400, label='N=400')
plt.loglog(x_fit_600, pdf_fit_600, label='N=600')
plt.loglog(x_fit_3600, pdf_fit_3600, label='N=3600')

plt.xlabel('s')
plt.ylabel('P(s; N)')
#plt.title('Fitted PDF of Power Law with Exponential Cutoff')
plt.legend()
#plt.savefig('2d_grid_graph_probability.png', dpi = 300)

plt.show()
```

```python
#  Scaling factors 
scaling_factors = [np.max(x_fit_100), np.max(x_fit_200), np.max(x_fit_400), np.max(x_fit_600), np.max(x_fit_3600)]

# Rescaling the avalanche_sizes data
avalanche_sizes_grid_100_scaled = (x_fit_100/scaling_factors[0]) 
avalanche_sizes_grid_200_scaled = x_fit_200/scaling_factors[1] 
avalanche_sizes_grid_400_scaled = x_fit_400/scaling_factors[2] 
avalanche_sizes_grid_600_scaled = x_fit_600/ scaling_factors[3] 
avalanche_sizes_grid_3600_scaled = x_fit_3600/ scaling_factors[4] 


pdf_fit_100_scaled = pdf_fit_100 * (x_fit_100)** 1.39
pdf_fit_200_scaled = pdf_fit_200 * ((x_fit_200 ) ** (1.39))
pdf_fit_400_scaled = pdf_fit_400 * ((x_fit_400) ** (1.39))
pdf_fit_600_scaled = pdf_fit_600  * ((x_fit_600) ** (1.39))
pdf_fit_3600_scaled = pdf_fit_3600  * ((x_fit_3600) ** (1.39))

# Plotting the collapsed data on a log-log plot
plt.figure()
plt.loglog(avalanche_sizes_grid_100_scaled, pdf_fit_100_scaled, label='N=100')
plt.loglog(avalanche_sizes_grid_200_scaled, pdf_fit_200_scaled, label='N=200')
plt.loglog(avalanche_sizes_grid_400_scaled, pdf_fit_400_scaled, label='N=400')
plt.loglog(avalanche_sizes_grid_600_scaled, pdf_fit_600_scaled, label='N=600')
plt.loglog(avalanche_sizes_grid_3600_scaled, pdf_fit_3600_scaled, label='N=3600')

plt.xlabel("$s/s_{max}}$")
plt.ylabel('$ s^{\u03C4} P(s;N)$')

#plt.title('Data Collapse')
plt.legend()
#plt.savefig("grid_data_collapse.png", dpi = 300)
plt.show()
```
