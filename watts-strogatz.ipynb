```python
import networkx as nx
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from main_numpy import NeuronModel
```

100 nodes

```python
# create watts-strogatz network
ws_network_100 = nx.watts_strogatz_graph(100, 10, 0.1) # 100 nodes, 10 neighborhs, and p=0.1
# add sink 
ws_network_100.add_node(-1)
# connect sink
for node in ws_network_100.nodes():
    if node != -1:
        if random.random() < 0.2:
            ws_network_100.add_edge(node, -1)

ws_model_100 = NeuronModel(ws_network_100)   

ws_data_100 = np.array(ws_model_100.run(50000))  # number of steps

# Calculate the unique avalanche sizes and their frequencies
avalanche_sizes_ws_100, frequencies_ws_100 = np.unique(ws_data_100, return_counts=True)
```

Plot simulated data

```python
def plot_scatter(avalanche_sizes, frequencies, label, color=None):
    plt.scatter(avalanche_sizes, frequencies, s=20, color=color, label=label)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('s')
    plt.ylabel('Frequency')
    plt.legend()
    plt.savefig(f'ws_{label}_scatter.png', dpi=300)
    plt.show()

data = [[avalanche_sizes_ws_100, frequencies_ws_100, 'N=100'],
        [avalanche_sizes_ws_200, frequencies_ws_200, 'N=200'], 
        [avalanche_sizes_ws_600, frequencies_ws_600, 'N=600'],
        [avalanche_sizes_ws_400, frequencies_ws_400, 'N=400'],
        [avalanche_sizes_ws_1600, frequencies_ws_1600, 'N=1600'],
        [avalanche_sizes_ws_3600, frequencies_ws_3600, 'N=3600'], 
        [avalanche_sizes_ws_4000, frequencies_ws_4000, 'N=4000']]

for avalanche_sizes, frequencies, label in data:
    plot_scatter(avalanche_sizes, frequencies, label)
```
